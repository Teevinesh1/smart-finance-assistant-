### Foundation Data Processing Functions

````python
# 🤖 AI Collaboration: Data Loading and Cleaning Function
# Ask AI to help you create a robust data loading function

def load_and_clean_transaction_data(file_path):
    """
    Load and clean transaction data for the Smart Finance Assistant

    🤖 AI Collaboration Prompt:
    "Create a function to load CSV transaction data with Date, Amount, Category,
    Description columns. Handle dollar signs in Amount, missing values, and
    data validation. Include clear business-focused error messages."

    Args:
        file_path: Path to CSV file or file object
    Returns:
        pandas.DataFrame: Cleaned transaction data
    """
    # Your AI-assisted implementation goes here
    # Read the CSV into a DataFrame
    df = pd.read_csv('april_2026_transactions.csv')

    # Make all column names lowercase and strip extra spaces
    # so "Amount" and "amount" and " Amount " all work the same
    df.columns = df.columns.str.strip().str.lower()

    # Drop completely blank rows
    df = df.dropna(how="all")

    # Remove $ signs and commas from the amount column so we can do maths
    # e.g. "$1,200.00" → "1200.00"
    df["amount"] = df["amount"].str.replace(r"[$,]", "", regex=True).str.strip()

    # Convert the amount from text to a real number
    df["amount"] = pd.to_numeric(df["amount"], errors="coerce")

    # Drop any rows where the amount still couldn't be read
    df = df.dropna(subset=["amount"])

    # Parse dates into proper date objects so we can sort and filter by date
    df["date"] = pd.to_datetime(df["date"], dayfirst=True)

    # Add a "day" column with just the day number (1–30)
    # — useful later for first half vs second half comparisons
    df["day"] = df["date"].dt.day

    # Label each row as "expense" or "refund"
    # Refunds are negative numbers in our CSV
    df["type"] = df["amount"].apply(lambda x: "refund" if x < 0 else "expense")

    # Add an absolute amount column (always positive) for totals and averages
    df["amount_abs"] = df["amount"].abs()

    # Tidy up category and description text — consistent capitalisation, no stray spaces
    df["category"]    = df["category"].str.strip().str.title()
    df["description"] = df["description"].str.strip()

    # Print a quick summary so we know the file loaded correctly
    print(f"  Loaded {len(df)} transactions")
    print(f"  Expenses : {(df['type'] == 'expense').sum()}")
    print(f"  Refunds  : {(df['type'] == 'refund').sum()}")
    print()
    print(df.to_string(index=False))

    return df

    pass

# Test your function
clean_data = load_and_clean_transaction_data("april_2026_transactions.csv")
print(clean_data)
````

````python
# 🤖 AI Collaboration: Spending Analysis Function
# Ask AI to help you create comprehensive spending analysis

def analyze_spending_patterns(df):
    """
    Analyze spending patterns and generate business insights

    🤖 AI Collaboration Prompt:
    "Create a function that analyses spending pattern and generate  actionable financial insight.
    It should calculate total spending, total spending per category, identify outliers,
    top expensive items, calculate 2 15 day period at the start and at the end in order to see which period spending is the most,
    calculate percentage of total spending expensive items are. Use simple language and use comments"

    """
    # Your AI-assisted implementation goes here
     # Work only with expenses — exclude refunds from spending analysis
    expenses = df[df["type"] == "expense"]

    line = "─" * 55

    # ── Total spending ────────────────────────────────────────
    total_spent = expenses["amount_abs"].sum()

    print()
    print("╔" + "═" * 55 + "╗")
    print("║{:^55}║".format("  SPENDING PATTERN ANALYSIS  "))
    print("╚" + "═" * 55 + "╝")

    print(f"\n{line}")
    print("  TOTAL SPENDING")
    print(line)
    print(f"  Total spent this month : ${total_spent:,.2f}")
    print(f"  Number of purchases    : {len(expenses)}")

    # ── Spending per category ─────────────────────────────────
    # Group all transactions by category and add up the amounts
    category_totals   = expenses.groupby("category")["amount_abs"].sum()
    category_counts   = expenses.groupby("category")["amount_abs"].count()
    category_averages = expenses.groupby("category")["amount_abs"].mean()

    print(f"\n{line}")
    print("  SPENDING BY CATEGORY")
    print(line)

    # Sort from highest to lowest so the biggest categories appear first
    for cat in category_totals.sort_values(ascending=False).index:
        total = category_totals[cat]
        count = category_counts[cat]
        avg   = category_averages[cat]
        pct   = (total / total_spent) * 100   # what % of total spend is this category
        bar   = "█" * int(pct / 5)            # simple bar — each block = 5%
        print(f"  {cat:<16} ${total:>8,.2f}  {pct:>5.1f}%  "
              f"avg ${avg:,.2f}/txn  ({int(count)} txn)  {bar}")

    # ── Outlier categories ────────────────────────────────────
    # A category is an outlier if it is more than 1.5x the average category spend
    # This helps flag categories where spending is disproportionately high
    avg_category_spend = category_totals.mean()
    outliers = category_totals[category_totals > avg_category_spend * 1.5]

    print(f"\n{line}")
    print("  OUTLIER CATEGORIES  (spending > 1.5x the category average)")
    print(line)

    if outliers.empty:
        print("  No outlier categories — spending is fairly balanced.")
    else:
        for cat, total in outliers.items():
            how_many_times = total / avg_category_spend
            print(f"  ⚠  {cat} — ${total:,.2f}  "
                  f"({how_many_times:.1f}x the average of ${avg_category_spend:,.2f})")

    # ── Top 5 most expensive items ────────────────────────────
    # Sort all individual transactions by amount and take the top 5
    top_5 = expenses.nlargest(5, "amount_abs")[
        ["date", "category", "amount_abs", "description"]
    ].reset_index(drop=True)

    print(f"\n{line}")
    print("  TOP 5 MOST EXPENSIVE PURCHASES")
    print(line)

    for i, row in top_5.iterrows():
        # Show what percentage of total spending this single item represents
        pct_of_total = (row["amount_abs"] / total_spent) * 100
        print(f"  {i+1}.  {row['date'].strftime('%d %b')}  "
              f"${row['amount_abs']:>8,.2f}  ({pct_of_total:.1f}% of total)  "
              f"{row['category']:<15}  {row['description']}")

    # ── 15-day period comparison ──────────────────────────────
    # Split the month into two halves and compare spending in each
    # This shows whether spending is heavier at the start or end of the month
    period_1 = expenses[expenses["day"] <= 15]   # days 1 to 15
    period_2 = expenses[expenses["day"] >  15]   # days 16 to end of month

    period_1_total = period_1["amount_abs"].sum()
    period_2_total = period_2["amount_abs"].sum()

    # Daily average for each period — fairer comparison since both have ~15 days
    period_1_daily_avg = period_1_total / 15
    period_2_daily_avg = period_2_total / 15

    # How much more (or less) was spent in period 2 vs period 1
    difference     = period_2_total - period_1_total
    diff_direction = "more" if difference > 0 else "less"

    print(f"\n{line}")
    print("  15-DAY PERIOD COMPARISON")
    print(line)
    print(f"  Period 1  (days  1–15)  ${period_1_total:>8,.2f}  "
          f"(daily avg ${period_1_daily_avg:,.2f})")
    print(f"  Period 2  (days 16–30)  ${period_2_total:>8,.2f}  "
          f"(daily avg ${period_2_daily_avg:,.2f})")
    print(f"\n  You spent ${abs(difference):,.2f} {diff_direction} in the second half of the month.")

    # ── Actionable insights ───────────────────────────────────
    print(f"\n{line}")
    print("  INSIGHTS & RECOMMENDATIONS")
    print(line)

    # Insight 1 — biggest category
    top_cat     = category_totals.idxmax()
    top_cat_pct = (category_totals[top_cat] / total_spent) * 100
    print(f"  • {top_cat} is your biggest expense at "
          f"${category_totals[top_cat]:,.2f} ({top_cat_pct:.1f}% of total spend).")

    # Insight 2 — period spending pattern
    if period_2_total > period_1_total:
        print(f"  • Your spending increases in the second half of the month. "
              f"Consider setting a mid-month budget limit.")
    else:
        print(f"  • Your spending is higher early in the month. "
              f"Good job slowing down in the second half!")

    # Insight 3 — saving tip for discretionary (non-essential) categories
    for cat in ["Entertainment", "Clothes"]:
        if cat in category_totals:
            saving = category_totals[cat] * 0.25
            print(f"  • Cutting {cat} by 25% would save you ~${saving:,.2f} this month.")

    # Insight 4 — overall savings potential
    print(f"  • Saving 20% of your total spend = ${total_spent * 0.20:,.2f} put away this month.")

    print(f"\n{line}\n")

    pass

# Test your function
analysis = analyze_spending_patterns(clean_data)
print(analysis)
````

````python
# 🤖 AI Collaboration: Business Insights Generator
# Ask AI to help create professional financial recommendations

def generate_financial_recommendations(analysis_data):
    """
    Generate actionable financial recommendations based on spending analysis

    🤖 AI Collaboration Prompt:
    "Based on spending analysis data, create professional financial
    recommendations. Include specific savings opportunities, spending
    pattern observations, and actionable advice formatted for a
    personal finance app user."

    Args:
        analysis_data: Dictionary with spending analysis results
    Returns:
        str: Formatted recommendations report
    """
    # Your AI-assisted implementation goes here
     # Work only with expenses — exclude refunds
    expenses = analysis_data[analysis_data["type"] == "expense"]
    refunds  = analysis_data[analysis_data["type"] == "refund"]

    # Calculate the key numbers we need for recommendations
    total_spent    = expenses["amount_abs"].sum()
    total_refunds  = refunds["amount_abs"].sum()
    net_spend      = total_spent - total_refunds

    # Spending per category
    category_totals = expenses.groupby("category")["amount_abs"].sum()

    # Split into first and second half of the month
    period_1_total = expenses[expenses["day"] <= 15]["amount_abs"].sum()
    period_2_total = expenses[expenses["day"] >  15]["amount_abs"].sum()

    # Top single purchase
    top_purchase = expenses.nlargest(1, "amount_abs").iloc[0]

    line      = "─" * 60
    thin_line = "·" * 60

    print()
    print("╔" + "═" * 60 + "╗")
    print("║{:^60}║".format("  💰  PERSONAL FINANCE RECOMMENDATIONS  💰  "))
    print("╚" + "═" * 60 + "╝")

    # ── 1. Spending snapshot ──────────────────────────────────
    # Give the user a quick summary of where they stand this month
    print(f"\n{line}")
    print("  📊  YOUR MONTH AT A GLANCE")
    print(line)
    print(f"  You spent ${total_spent:,.2f} this month across {len(expenses)} purchases.")
    print(f"  After ${total_refunds:,.2f} in refunds, your net spend was ${net_spend:,.2f}.")

    # Work out the daily average so we can give a relatable benchmark
    daily_avg = total_spent / 30
    print(f"  That works out to roughly ${daily_avg:,.2f} per day.")

    # ── 2. Spending pattern observations ─────────────────────
    # Highlight what the data tells us about how the user spends
    print(f"\n{line}")
    print("  🔍  SPENDING PATTERN OBSERVATIONS")
    print(line)

    # Observation 1 — which half of the month costs more
    if period_2_total > period_1_total:
        diff = period_2_total - period_1_total
        print(f"  • End-of-month surge: You spent ${diff:,.2f} more in the second")
        print(f"    half of the month (${period_2_total:,.2f}) than the first (${period_1_total:,.2f}).")
        print(f"    This is common after payday — try setting a weekly limit")
        print(f"    to keep spending even across the month.")
    else:
        diff = period_1_total - period_2_total
        print(f"  • Strong start discipline: You spent ${diff:,.2f} more in the")
        print(f"    first half (${period_1_total:,.2f}) but slowed down in the second")
        print(f"    half (${period_2_total:,.2f}). Great spending discipline!")

    # Observation 2 — biggest category as a share of total
    top_cat       = category_totals.idxmax()
    top_cat_total = category_totals[top_cat]
    top_cat_pct   = (top_cat_total / total_spent) * 100
    print(f"\n  • {top_cat} dominates your spending at ${top_cat_total:,.2f}")
    print(f"    ({top_cat_pct:.1f}% of everything you spent this month).")

    # Observation 3 — flag if any single purchase is more than 10% of total
    top_pct = (top_purchase["amount_abs"] / total_spent) * 100
    if top_pct > 10:
        print(f"\n  • Large single purchase: {top_purchase['description']}")
        print(f"    cost ${top_purchase['amount_abs']:,.2f} — that's {top_pct:.1f}% of your")
        print(f"    entire monthly spend in one transaction.")

    # ── 3. Savings opportunities ──────────────────────────────
    # Concrete, specific actions the user can take right now
    print(f"\n{line}")
    print("  💡  SAVINGS OPPORTUNITIES")
    print(line)

    saving_number = 1   # used to number each tip

    # Tip for each discretionary (non-essential) category
    # Entertainment and Clothes are wants, not needs — easier to cut
    for cat in ["Entertainment", "Clothes"]:
        if cat in category_totals:
            cat_total    = category_totals[cat]
            save_25      = cat_total * 0.25   # saving from a 25% cut
            cat_pct      = (cat_total / total_spent) * 100
            print(f"\n  {saving_number}. Reduce {cat} spending")
            print(f"     You spent ${cat_total:,.2f} on {cat} ({cat_pct:.1f}% of total).")
            print(f"     Cutting this by 25% saves ~${save_25:,.2f}/month")
            print(f"     — that's ~${save_25 * 12:,.2f} over a full year.")
            saving_number += 1

    # Tip — build an emergency fund from net spend
    emergency_fund_contribution = net_spend * 0.10   # 10% of net spend
    print(f"\n  {saving_number}. Start an emergency fund")
    print(f"     Setting aside just 10% of your net spend (${emergency_fund_contribution:,.2f}/month)")
    print(f"     builds a ${emergency_fund_contribution * 12:,.2f} safety net over 12 months.")
    saving_number += 1

    # Tip — 50/30/20 budget check
    # 50% needs, 30% wants, 20% savings — a classic personal finance framework
    needs_budget   = net_spend * 0.50
    wants_budget   = net_spend * 0.30
    savings_budget = net_spend * 0.20
    needs_actual   = sum(category_totals.get(c, 0) for c in ["Food", "Transport"])
    wants_actual   = sum(category_totals.get(c, 0) for c in ["Entertainment", "Clothes"])

    print(f"\n  {saving_number}. Check your 50/30/20 budget balance")
    print(f"     A healthy budget splits net spend into:")
    print(f"     Needs  (50%)  target ${needs_budget:,.2f}  —  actual ${needs_actual:,.2f}")
    print(f"     Wants  (30%)  target ${wants_budget:,.2f}  —  actual ${wants_actual:,.2f}")
    print(f"     Savings(20%)  target ${savings_budget:,.2f}")

    # Flag if wants are over budget
    if wants_actual > wants_budget:
        over_by = wants_actual - wants_budget
        print(f"     ⚠  Your 'wants' spending is ${over_by:,.2f} over the recommended limit.")

    # ── 4. Actionable next steps ──────────────────────────────
    # Clear, numbered things the user can do this week
    print(f"\n{line}")
    print("  ✅  YOUR ACTION PLAN FOR NEXT MONTH")
    print(line)
    print(f"  1. Set a daily spending limit of ${daily_avg * 0.85:,.2f}")
    print(f"     (15% less than your current ${daily_avg:,.2f} daily average).")
    print(f"  2. Review your {top_cat} purchases — look for cheaper alternatives.")
    print(f"  3. Do a mid-month check-in on day 15 to stay on track.")
    print(f"  4. Automate ${savings_budget:,.2f} into savings on payday so it")
    print(f"     never gets spent.")

    print(f"\n{thin_line}")
    print(f"  Tip: Small consistent changes beat big one-off cutbacks.")
    print(f"  Even saving ${savings_budget:,.2f}/month = ${savings_budget * 12:,.2f} by this time next year.")
    print(f"{thin_line}\n")
    pass

# Test your function
analysis = clean_data
recommendations = generate_financial_recommendations(analysis)
print(analysis)
````

````python
# 🤖 AI Collaboration: Document Retrieval Setup
# Ask AI to help you set up RAG for financial documents

from hands_on_ai import rag

def setup_financial_rag():
    """
    Set up RAG system for financial documents and transaction data

    🤖 AI Collaboration Prompt:
    "Help me set up a RAG system that can retrieve information from
    financial documents, budgeting guides, and transaction summaries
    to answer user questions about personal finance."
    """
    # Your AI-assisted RAG setup goes here
    # Include document ingestion, query processing
    # Create our own finance knowledge document
    finance_document = """
    The 50/30/20 rule splits income into three parts.
    50 percent goes to needs like rent food and transport.
    30 percent goes to wants like entertainment and clothes.
    20 percent goes to savings and investments.

    To stop overspending on entertainment set a fixed monthly budget.
    Use the 24 hour rule before any non essential purchase.
    Cancel unused subscriptions and find free alternatives.

    To build savings pay yourself first before spending on anything else.
    Set up automatic transfers to savings on payday.
    Start with saving 10 percent and increase gradually.

    To reduce food spending meal prep on weekends to avoid takeaways.
    Make a shopping list before going to the store.
    Cook at home at least 5 days a week to save money.

    To manage transport costs use public transport when possible.
    Carpool with colleagues to split fuel costs.
    Walk or cycle for short distances to save on fuel.

    An emergency fund should cover 3 to 6 months of expenses.
    Keep emergency funds in a separate savings account.
    Only use emergency funds for genuine emergencies.
    """

    # Save the finance document as a text file
    with open("finance_guide.txt", "w") as f:
        f.write(finance_document)
    print("📄 Finance document created!")

    # Split the document into small chunks for searching
    chunks = rag.chunk_text(finance_document)
    sources = ["finance_guide.txt"] * len(chunks)
    print(f"✅ Document split into {len(chunks)} chunks!")

    return chunks, sources

def ask_financial_question(question, chunks, sources):
    # Search through chunks manually to find the most relevant answer
    # Since get_embeddings needs Ollama, we do simple keyword matching instead
    question_words = question.lower().split()

    # Score each chunk by how many question words it contains
    scored_chunks = []
    for chunk, source in zip(chunks, sources):
        score = sum(1 for word in question_words if word in chunk.lower())
        scored_chunks.append((score, chunk, source))

    # Sort by score and take the top 3 most relevant chunks
    scored_chunks.sort(reverse=True)
    top_chunks = scored_chunks[:3]

    print(f"\n❓ Question: {question}\n")
    print("📚 Most relevant information found:")
    print("─" * 50)
    for score, chunk, source in top_chunks:
        print(f"\n  {chunk}")
    print("─" * 50)

# Set up the RAG system
chunks, sources = setup_financial_rag()
pass

# Test your RAG system
answer = ask_financial_question(
    "What's a good budgeting strategy for someone who overspends on entertainment?",
    chunks, sources
)
````

## Custom Financial Tools

````python
# 🤖 AI Collaboration: Custom Tool Development
# Ask AI to help you create useful financial calculators

from hands_on_ai import agent

def create_savings_calculator_tool(current_savings: float, monthly_contribution: float, target_amount: float) -> str:
    # Savings goal calculator that takes current savings, monthly contribution
    # and target amount then calculates how long it will take to reach the goal
    # AI Collaboration Prompt: Create a savings goal calculator function that takes current savings,
    # monthly contribution, and target amount, then calculates time to reach
    # goal. Format output for user friendly display.

    # Validate inputs — all values must be positive numbers
    if current_savings < 0 or monthly_contribution <= 0 or target_amount <= 0:
        return "❌ Please enter positive values for all fields."

    # Check if goal is already reached
    if current_savings >= target_amount:
        return "✅ You have already reached your savings goal!"

    # Calculate how much more is needed
    amount_needed = target_amount - current_savings

    # Calculate how many months it will take to reach the goal
    months_needed = amount_needed / monthly_contribution

    # Convert months to years and remaining months for easier reading
    years  = int(months_needed // 12)
    months = int(months_needed % 12)

    # Format the result in a user friendly way
    line = "─" * 45
    result = (
        f"\n{line}\n"
        f"  💰 SAVINGS GOAL CALCULATOR\n"
        f"{line}\n"
        f"  Current savings     : ${current_savings:,.2f}\n"
        f"  Monthly contribution: ${monthly_contribution:,.2f}\n"
        f"  Target amount       : ${target_amount:,.2f}\n"
        f"  Amount still needed : ${amount_needed:,.2f}\n"
        f"{line}\n"
        f"  ⏱  Time to reach goal: {years} years and {months} months\n"
        f"{line}\n"
    )
    return result

# Register your tool with the agent system
agent.register_tool(
    name="savings_calculator",
    description="Calculates how long it will take to reach a savings goal based on current savings and monthly contributions",
    function=create_savings_calculator_tool
)
print("✅ Savings calculator tool registered!\n")

# Test the tool directly without the agent
result = create_savings_calculator_tool(
    current_savings=500,
    monthly_contribution=200,
    target_amount=5000
)
print(result)

# Test your tool with different savings scenarios
print("Test 1 — Standard savings goal:")
print(create_savings_calculator_tool(500, 200, 5000))

print("Test 2 — Already reached goal:")
print(create_savings_calculator_tool(6000, 200, 5000))

print("Test 3 — Large savings goal:")
print(create_savings_calculator_tool(1000, 500, 20000))
````

````python
import pandas as pd

# =========================================================
# 1. CREATE TEST DATASETS
# =========================================================

def create_test_datasets():
    """
    Create realistic test datasets for finance assistant
    """

    # ---------------- Normal data ----------------
    normal = pd.DataFrame({
        "date": ["01/04/2026", "03/04/2026", "05/04/2026"],
        "amount": ["$45.50", "$120.00", "$250.00"],
        "category": ["Food", "Transport", "Shopping"],
        "description": ["Woolworths", "Uber", "JB Hi-Fi"]
    })

    # ---------------- Edge cases ----------------
    edge_cases = pd.DataFrame({
        "date": ["02/04/2026", "08/04/2026", "10/04/2026"],
        "amount": ["-$50.00", "$0.00", "$9999.99"],
        "category": ["Refund", "Food", "Electronics"],
        "description": ["Amazon refund", "", "MacBook Pro"]
    })

    # ---------------- Invalid data ----------------
    invalid = pd.DataFrame({
        "date": ["bad_date", "", "15/04/2026"],
        "amount": ["abc", "twenty", "$100.00"],
        "category": ["Food", None, "Transport"],
        "description": ["bad", "missing", "valid"]
    })

    # ---------------- High spending ----------------
    high_spending = pd.DataFrame({
        "date": ["01/04/2026", "10/04/2026"],
        "amount": ["$2000.00", "$1500.00"],
        "category": ["Entertainment", "Travel"],
        "description": ["Concert", "Flight"]
    })

    return {
        "normal": normal,
        "edge": edge_cases,
        "invalid": invalid,
        "high": high_spending
    }


# =========================================================
# 2. TEST DATA LOADING
# =========================================================

def test_data_loading_function():
    print("🧪 Testing data loading function...")

    datasets = create_test_datasets()

    # Save normal dataset
    datasets["normal"].to_csv("test_normal.csv", index=False)

    df = load_and_clean_transaction_data("test_normal.csv")

    # Core checks (robust, NOT fixed row count)
    assert len(df) > 0, "Should return data after cleaning"
    assert "amount" in df.columns, "Amount column missing"
    assert pd.api.types.is_float_dtype(df["amount"]), "Amount must be float"
    assert df["amount"].isna().sum() == 0, "No NaN values allowed"

    print("✔ Data loading tests passed")


# =========================================================
# 3. TEST SPENDING ANALYSIS
# =========================================================

def test_spending_analysis():
    print("🧪 Testing spending analysis...")

    df = pd.DataFrame({
        "date": pd.to_datetime(["2026-04-01", "2026-04-05", "2026-04-10"]),
        "amount": [100, 200, -50],
        "category": ["Food", "Transport", "Refund"],
        "description": ["A", "B", "C"]
    })

    df["day"] = df["date"].dt.day
    df["type"] = df["amount"].apply(lambda x: "refund" if x < 0 else "expense")
    df["amount_abs"] = df["amount"].abs()

    expenses = df[df["type"] == "expense"]

    total = expenses["amount_abs"].sum()

    # Refund excluded
    assert total == 300, "Refund should be excluded"

    # Category logic
    cat_totals = expenses.groupby("category")["amount_abs"].sum()

    assert cat_totals["Food"] == 100
    assert cat_totals["Transport"] == 200

    # Percent check
    assert round((cat_totals.sum() / total) * 100, 2) == 100.00

    print("✔ Spending analysis tests passed")


# =========================================================
# 4. TEST BUSINESS INSIGHTS
# =========================================================

def test_business_insights():
    print("🧪 Testing business insights...")

    df = pd.DataFrame({
        "amount": [300, 200, 100, 50],
        "category": ["Entertainment", "Food", "Entertainment", "Transport"]
    })

    df["amount_abs"] = df["amount"]

    cat_totals = df.groupby("category")["amount_abs"].sum()

    # Highest spending category
    assert cat_totals.idxmax() == "Entertainment"

    # Savings calculation
    savings = cat_totals["Entertainment"] * 0.25
    assert savings == 100.0

    # Logic check
    assert "Food" in cat_totals.index

    print("✔ Business insights tests passed")


# =========================================================
# 5. RUN ALL TESTS
# =========================================================

print("🔍 COMPREHENSIVE TESTING SUITE")
print("=" * 40)

try:
    create_test_datasets()
    test_data_loading_function()
    test_spending_analysis()
    test_business_insights()

    print("\n✅ ALL TESTS PASSED — FINANCE ASSISTANT IS WORKING")

except AssertionError as e:
    print(f"\n❌ TEST FAILED: {e}")

except Exception as e:
    print(f"\n⚠️ ERROR OCCURRED: {e}")
````
